```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
y_true = np.array([1,0,1,0,1,0,0,1])
y_prob = np.array([0.9,0.8,0.7,0.6,0.5,0.4,0.3,0.2])
```

```python
idx = np.argsort(y_prob)[::-1]
idx
```

```python
y_sorted=y_true[idx]
y_sorted
```

```python
cum_pos = np.cumsum(y_sorted)
cum_pos
```

```python
total_pos = np.sum(y_sorted)
total_pos
```

```python
cap_y = cum_pos / total_pos
cap_y
```

```python
cap_x=np.arange(1, len(y_true)+1)/len(y_true)
cap_x
```

```python
random_y=cap_x
```

```python
perfect_y=np.sort(y_true)[::-1].cumsum()/total_pos
```

```python
plt.figure(figsize=(8,6))

plt.plot(cap_x, cap_y, label="Random Forest Model", linewidth=2)
plt.plot(cap_x, random_y, linestyle="--", label="Random Model")
plt.plot(cap_x, perfect_y, linestyle="--", label="Perfect Model")

plt.xlabel("Cumulative % of Population")
plt.ylabel("Cumulative % of Positives Captured")
plt.title("CAP Curve - Random Forest Classifier")
plt.legend()
plt.grid()
plt.show()
```
